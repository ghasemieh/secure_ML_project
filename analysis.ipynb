```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("record.csv")
y_true = data["y_true"]
y_diff = data["y_diff"]
y_adv = data["y_adv"]
```

```python
MAX_DATA = 90
f, (ax1) = plt.subplots(nrows = 1, ncols = 1)
f.set_figheight(6)
f.set_figwidth(20)
ax1.set_xlabel('Iteration')
ax1.set_ylabel('Steering Angle')
ax1.plot(y_true[40:MAX_DATA], 'k-', label='y_true', linewidth=3)
ax1.plot(y_adv[40:MAX_DATA], 'b-', label='y_adv', linewidth=3)
ax1.legend()
ax1.hlines(0, 0, 50-1, color='red', linestyles='dashed')
plt.show()
```
